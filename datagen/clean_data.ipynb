```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from IPython import display
from IPython.display import clear_output
```

```python
import sys, os
from PIL import Image
import shutil

def display_image(path):
    image = mpimg.imread(path)
    plt.imshow(image)
    plt.title(os.path.basename(path))
    plt.show()
    

def saveWhere(path, save_male='./male', save_female='./female', save_reject='./reject'):
    print('Enter 1 if male, 2 if female, 3 to reject it, 4 to end the process\n', flush=True)
    num = input()
    filename = os.path.basename(path)
    if(num=='1'):
        savepath = save_male + '/' + filename
    elif(num=='2'):
        savepath = save_female + '/' +filename
    elif(num=='3'):
        savepath = save_reject + '/' +filename
    else:
        return 4
        
    print(f'path = {path}\nsave_path = {savepath}')
    shutil.move(path, savepath)
        
    
```

```python
def organize_dataset(datadir='./downloaded_images', save_male='./male', save_female='./female', save_reject='./reject'):
    if(not os.path.exists(save_male)):
        os.mkdir(save_male)
    if(not os.path.exists(save_female)):
        os.mkdir(save_female)
    if(not os.path.exists(save_reject)):
        os.mkdir(save_reject)

    files = os.listdir(datadir)
    for f in files:
        if(f.endswith(".jpg") or f.endswith(".png")):
            impath = datadir + '/' + f
            clear_output(wait=True)
            display_image(impath)
            status = saveWhere(impath, save_male, save_female)
            if(status==4):
                clear_output(wait=False)
                print('Exiting data organization! You can run this function to process remaining data.\n')
                return None
    print("Finished processing the dataset!\n")
            
        
```

```python
organize_dataset()
```
